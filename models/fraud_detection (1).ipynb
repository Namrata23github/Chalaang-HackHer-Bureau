```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# Define the dimensions of the input data
input_dim =  robust_scaled_data.shape[1]

# Define the architecture of the autoencoder
def build_autoencoder(input_dim):
    # Encoder
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(64, activation='relu')(input_layer)
    encoded = Dense(32, activation='relu')(encoded)

    # Latent space
    encoded = Dense(16, activation='relu')(encoded)

    # Decoder
    decoded = Dense(32, activation='relu')(encoded)
    decoded = Dense(64, activation='relu')(decoded)
    decoded = Dense(input_dim, activation='sigmoid')(decoded)  # Sigmoid for binary data, linear for continuous data

    # Autoencoder model
    autoencoder = Model(input_layer, decoded)
    autoencoder.compile(optimizer='adam', loss='mse')  # Mean Squared Error loss for reconstruction

    return autoencoder

# Build the autoencoder model
autoencoder = build_autoencoder(input_dim)

# Train the autoencoder

from sklearn.model_selection import train_test_split

# Assuming X contains your transaction data and y contains corresponding labels (if available)
# Split the data into training and testing sets
X_train, X_test = train_test_split(robust_scaled_data, test_size=0.2, random_state=42)
# Assuming you have a dataset X_train containing your transaction data
autoencoder.fit(X_train, X_train, epochs=10, batch_size=32, shuffle=True)

# Make predictions on the training data
reconstructed_data = autoencoder.predict(X_train)

# Calculate reconstruction error
mse = np.mean(np.power(X_train - reconstructed_data, 2), axis=1)

# Set a threshold for anomaly detection
threshold = np.mean(mse) + 2 * np.std(mse)  # Adjust multiplier for desired sensitivity

# Detect anomalies
anomalies = np.where(mse > threshold)[0]
print("Indices of anomalies:", anomalies)
```

```python
# Make predictions on the test data
reconstructed_data = autoencoder.predict(X_test)

# Calculate reconstruction error
mse = np.mean(np.power(X_test - reconstructed_data, 2), axis=1)

# Set a threshold for anomaly detection
threshold = np.mean(mse) + 2 * np.std(mse)  # Adjust multiplier for desired sensitivity

# Detect anomalies
anomalies = np.where(mse > threshold)[0]
print("Indices of anomalies:", anomalies)
```
